```

#1 different functions of string in python
 
print("STRING FUNCTIONS")
 
name="Sayyed Nurjahan"
 
print(name)
 
#swapcase()
 
name=name.swapcase()
 
print("swapcase= ",name)
 
#upper case()
 
name=name.upper()
 
print("uppercase= ",name)
 
#lowercase()
 
name=name.lower()
 
print("lowercase= ",name)
 
#title()
 
name=name.title()
 
print("title= ",name)
 
#islower()
 
name=name.islower()
 
print("islower= ",name)
 
print("\n")
 
#2 different functions of list in python
 
a=[1,2,3,"nuru",5,'a','b']
 
print("LIST FUNCTIONS")
 
#type
 
b=type(a)
 
print("type= ",b)
 
#index()
 
b=a.index(2)
 
print("index= ",b)
 
#remove()
 
b=a.remove(2)
 
print("remove= ",a)
 
#change function 
 
a[4]=4
 
print(a)
 
#append
 
a.append(5)
 
print("append= ",a)
 
#reverse
 
a.reverse()
 
print("reverse= ",a)
 
print("\n")
 
#different functions of dictionary 
 
print("DICTIONARY  FUNCTIONS")
 
data={1:'a',2:'b',3:'c',4:'d'}
 
print(data)
 
print(data[4])
 
#get()
 
print(data.get(1))
 
#pop()
 
data.pop(4)
 
print(data)
 
#adding two list as a dictionary 
 
keys=[1,2,3]
 
values=['a','b','c']
 
print(keys,values)
 
data=dict(zip(keys,values))
 
print(data)
 
#delete()
 
del(data[2])
 
print(data)
 
#sub dictionary 
 
data={1:'a',2:'b',3:['c','d']}
 
print(data)
 
print(data[1])
 
print(data[3][1])
```
